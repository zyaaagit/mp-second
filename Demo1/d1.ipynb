```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset1 = pd.read_excel("附件.xlsx")
data = pd.DataFrame(dataset1)
```

```python
data_lx1 = data[data["类型"] == "高钾"]
data_lx2 = data[data["类型"] == "铅钡"]

data_Ws1 = data[data["纹饰"] == "A"]
data_Ws2 = data[data["纹饰"] == "B"]
data_Ws3 = data[data["纹饰"] == "C"]

data_ys1 = data[data["颜色"] == "黑"]
data_ys2 = data[data["颜色"] == "蓝绿"]
data_ys3 = data[data["颜色"] == "绿"]
data_ys4 = data[data["颜色"] == "浅蓝"]
data_ys5 = data[data["颜色"] == "浅绿"]
data_ys6 = data[data["颜色"] == "深蓝"]
data_ys7 = data[data["颜色"] == "深绿"]
data_ys8 = data[data["颜色"] == "紫"]
```

```python
count_lx1 = data_lx1["表面风化"].value_counts(normalize=True)
count_lx2 = data_lx2["表面风化"].value_counts(normalize=True)

count_ws1 = data_Ws1["表面风化"].value_counts(normalize=True)
count_ws2 = data_Ws2["表面风化"].value_counts(normalize=True)
count_ws3 = data_Ws3["表面风化"].value_counts(normalize=True)

count_ys1 = data_ys1["表面风化"].value_counts(normalize=True)
count_ys2 = data_ys2["表面风化"].value_counts(normalize=True)
count_ys3 = data_ys3["表面风化"].value_counts(normalize=True)
count_ys4 = data_ys4["表面风化"].value_counts(normalize=True)
count_ys5 = data_ys5["表面风化"].value_counts(normalize=True)
count_ys6 = data_ys6["表面风化"].value_counts(normalize=True)
count_ys7 = data_ys7["表面风化"].value_counts(normalize=True)
count_ys8 = data_ys8["表面风化"].value_counts(normalize=True)
```

```python
count_ys1
```

```python
plt.rcParams['font.sans-serif'] = ['SimHei']  # 黑体
plt.rcParams['axes.unicode_minus'] = False    # 解决无法显示符号的问题
sns.set(font='SimHei', font_scale=1.0)        # 解决Seaborn中文显示问题
x_lx = ["高钾","铅钡"]
fh_1 = [count_lx1["风化"],count_lx2["风化"]]
fh_2 = [count_lx1["无风化"],count_lx2["无风化"]]
fig, ax = plt.subplots(figsize=(5, 5))
# 先创建一根柱子，显示第一种产品的销量
ax.bar(x_lx, fh_1, color="red", width=0.8,label="风化")
# 第二根柱子“堆积”在第一根柱子上方，通过'bottom'调整，显示第二种产品的销量
ax.bar(x_lx, fh_2, color="blue", bottom=fh_1, width=0.8, label="无风化")
ax.set_title("类型与风化的关系图", fontsize=15)
ax.set_xlabel("玻璃类型")
ax.set_ylabel("百分比")
ax.legend()
```

```python
plt.rcParams['font.sans-serif'] = ['SimHei']  # 黑体
plt.rcParams['axes.unicode_minus'] = False    # 解决无法显示符号的问题
sns.set(font='SimHei', font_scale=1.0)        # 解决Seaborn中文显示问题
x_lx = ["A","B", "C"]
fh_1 = [count_ws1["风化"],count_ws2["风化"], count_ws3["风化"]]
fh_2 = [count_ws1["无风化"],0, count_ws3["无风化"]]

fig, ax = plt.subplots(figsize=(5, 5))
# 先创建一根柱子，显示第一种产品的销量
ax.bar(x_lx, fh_1, color="red", width=0.8,label="风化")
# 第二根柱子“堆积”在第一根柱子上方，通过'bottom'调整，显示第二种产品的销量
ax.bar(x_lx, fh_2, color="blue", bottom=fh_1, width=0.8, label="无风化")
ax.set_title("纹饰与风化的关系图", fontsize=15)
ax.set_xlabel("纹饰")
ax.set_ylabel("百分比")
ax.legend()
```
