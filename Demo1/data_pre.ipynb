```python
#对表单1进行缺失值处理
import pandas as pd
```

```python
dataset1 = pd.read_excel("附件.xlsx",sheet_name = "合并")
data = pd.DataFrame(dataset1)
data = data.fillna(0)
```

```python
data.head(12)
```

```python
ych = list()
for i in data.index:
    b = 0
    for j in range(6,19):
        b += data.iloc[i,j]
    if b >= 105 or b <= 85:
        ych.append(i)
```

```python
data = data.drop(ych)
```
