```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.mlab as mlab 
```

```python
dataset2 = pd.read_excel("Data.xlsx")
data = pd.DataFrame(dataset2)
data = data.drop(data.columns[0], axis = 1)
```

```python
data_lx1 = data[data["类型"] == "高钾"][data["表面风化"] == "风化"]
data_lx2 = data[data["类型"] == "高钾"][data["表面风化"] == "无风化"]
data_lx3 = data[data["类型"] == "铅钡"][data["表面风化"] == "风化"]
data_lx4 = data[data["类型"] == "铅钡"][data["表面风化"] == "无风化"]
```

```python
plt.figure(figsize=(40,20))
a = 0
for i in range(6,18):
    xlab = data_lx1.columns[i]
    hxcf_x = pd.DataFrame({"风化高钾":data_lx1.iloc[:,i],"无风化高钾":data_lx2.iloc[:,i],"风化铅钡":data_lx3.iloc[:,i],"无风化铅钡":data_lx4.iloc[:,i]})
    plt.subplot(3,4,a+1)
    a += 1
    #plt.subplots(dpi=1080,facecolor='w')# 设置画布大小，分辨率，和底色
    hxcf_x.boxplot()
    plt.xlabel(xlab,fontsize = 16) # 我们设置横纵坐标的标题。
    plt.tick_params(labelsize = 16) 
```
